#Clean Dataset

````python
#Checking duplicates
dup=data.duplicated().any()
print("Is there any duplicate?",dup)
````

````python
#checking Null Values
````

````python
data.isnull()
````

````python
data.isnull().sum()
````

````python
#drop null values
data.dropna(axis=0,how='any',inplace=True)
````

````python
data.shape
````

````python
data.isnull().sum()
````

````python
data.head()
````

````python
data.rename(columns={'Rotten Tomatoes':'RottenTomatoes'},inplace=True)
data.head()
````

````python
def sr(x):
  x=x.split("/")
  x=float(x[0]) / float(x[1])
  return float(x)
data['RottenTomatoes']=data['RottenTomatoes'].apply(sr)
  
````

````python
data.tail()
````

````python
data['RottenTomatoes']=data['RottenTomatoes'].apply(lambda x:x*100)
data.tail()
````

````python
data['RottenTomatoes'] = data['RottenTomatoes'].apply(int)
data.tail()
````

**How many movies from this dataset were made in 2010 and imdb rating more than equal to 5**

````python
data.columns
````

````python
def sr(x):
  x=x.split("/")
  x=float(x[0]) / float(x[1])
  return float(x)
data['IMDb']=data['IMDb'].apply(sr)
data
````

````python
data['IMDb']=data['IMDb'].apply(lambda x: x*10)
data.head()
````

````python
data['IMDb']=data['IMDb'].apply(float)
data.head()
````

````python
data.IMDb.dtype
````

````python
data[(data['IMDb']>=5) &(data['Year']==2010)].count()
````

**How many different age categories are there**

````python
agegroup=data.groupby('Age')['Age'].count()
agegroup
````

````python
plt.plot(agegroup,linestyle='dashed')
plt.xlabel("Age restriction")
plt.ylabel("No of movies")
plt.show()
````
