```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading a dataset
data=pd.read_excel("Spotify.xlsx")
data
```

```python
#information of the dataset
data.info()
```

```python
#Rename the column
data.rename(columns={'year released':'year','top genre':'top_genre'},inplace=True)
data.columns
```

```python
#Display the visulization how energetic the song is vs How easy it is to dance to the song
plt.figure(figsize=(16,9))
sns.scatterplot(x='nrgy',y='dnce',data=data,hue='artist type')
plt.title("Energy Vs Dancebility")
plt.xlabel("Energy - How energetic the song is")
plt.ylabel("Danceability - How easy it is to dance to the song")
plt.show()
```

```python
data.head(1)
```

```python
#Display the visulization how loud  the song is vs How easy it is to dance to the song
plt.figure(figsize=(16,9))
sns.scatterplot(y='dnce',x='dB',data=data,hue='artist type')
plt.title("dancebility vs Loud the song ")
plt.ylabel("Danceability - How easy it is to dance to the song")
plt.xlabel("Decibel - How loud the song is")
plt.show()
```

```python
data.head(1)
```

```python
#Display the visulization how much beats per minute in a duration
plt.figure(figsize=(16,9))
sns.scatterplot(x='bpm',y='dur',data=data,color='red')
plt.title("how much beats per minute in a duration")
plt.xlabel("beats per minute")
plt.ylabel("duration of the song")
plt.show()
```

```python
#Display the visulization type of artist
plt.figure(figsize=(16,9))
sns.countplot(x='artist type',data=data)
plt.title("type of artist ")
plt.xlabel("artist type")
plt.ylabel("Count")
plt.show()
```

```python
#Group  the added column
data.groupby('added')['added'].count()
```

```python
#Display the visulization how many songs added in spotity
plt.figure(figsize=(35,10))
sns.countplot(x='added', data=data)
plt.xlabel("Added of the year song")
plt.title("How many songs added in spotify")
plt.show()
```

```python
#Group the year column
data.groupby('year')['year'].count()
```

```python
#Display the visulization how many songs released per year
plt.figure(figsize=(16,9))
sns.countplot(x='year',data=data)
plt.xlabel("year released of the song")
plt.title("How many songs released per year")
plt.show()
```

```python
#Sort the values in descending order How positive the mood of the song is
data.sort_values('val',ascending=False)
```

```python
#How acoustic the song is arrange in descending order

data.sort_values('acous',ascending=False)
```
